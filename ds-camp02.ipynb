## 실습1. 거품 정렬(@astrachan2003bubble)

```python
data = [0,3,1,2,5,10,5,33,1]
print(data)
for i in range(len(data)-1,0,-1):
    for j in range(i):
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
print(data)
```

## 실습2. 선택 정렬(@mishra2008selection)

```python
data = [0,3,1,2,5,10,5,33,1]
print(data)
for i in range(len(data)):
    idx = i
    for j in range(i+1, len(data)):
        if data[idx] > data[j]: 
            idx = j
    data[i], data[idx] = data[idx], data[i]
print(data)
```

## 실습3. 삽입 정렬(@sodhi2013enhanced)

```python
data = [0,3,1,2,5,10,5,33,1]
print(data)
for end in range(1, len(data)):
    for i in range(end, 0, -1):
        if data[i - 1] > data[i]:
            data[i - 1], data[i] = data[i], data[i - 1]
print(data)
```
