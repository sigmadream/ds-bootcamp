## 실습4. 거품 정렬(@astrachan2003bubble)

```python
def bubble_sort(data = [0,3,1,2,5,10,5,33,1]):    
    print("Input : ", data)
    for i in range(len(data)-1,0,-1):
        for j in range(i):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data
print("Output : ", bubble_sort())
```

## 실습5. 선택 정렬(@mishra2008selection)

```python
def selection_sort(data = [0,3,1,2,5,10,5,33,1]):
    print("Input : ", data)
    for i in range(len(data)):
        idx = i
        for j in range(i+1, len(data)):
            if data[idx] > data[j]: 
                idx = j
        data[i], data[idx] = data[idx], data[i]
    return data
print("Output : ", selection_sort())
```

## 실습6. 삽입 정렬(@sodhi2013enhanced)

```python
def insertion_sort(data = [0,3,1,2,5,10,5,33,1]):
    print("Input : ", data)
    for end in range(1, len(data)):
        for i in range(end, 0, -1):
            if data[i - 1] > data[i]:
                data[i - 1], data[i] = data[i], data[i - 1]
    return data
print("Output : ", insertion_sort())
```

## 실습7. 퀵 정렬(@hoare1962quicksort)

```python
def quick_sort(l=[3, 0, 1, 8, 7, 2, 5, 4, 6, 9]):
    if len(l) <= 1:
        return l
    pivot = l[len(l) // 2]
    lesser_arr, equal_arr, greater_arr = [], [], []
    for num in l:
        if num < pivot:
            lesser_arr.append(num)
        elif num > pivot:
            greater_arr.append(num)
        else:
            equal_arr.append(num)
    return quick_sort(lesser_arr) + equal_arr + quick_sort(greater_arr)

print(quick_sort())
```

## 실습8(재귀) 문자열 뒤집기

```python
def reverse_recursive(s="HelloWorld"):
    if not s: return []
    return [s[-1]] + reverse_recursive(s[:-1])
print("".join(reverse_recursive()))
```

## 실습9(재귀). 리스트에서 원소 가져오기

```python
def take_recursive(l=[1,2,3,4,5,6],n=2):
    if n <= 0 or not l: return []
    return [l[0]] + take_recursive(l[1:], n-1)
print(take_recursive(n=4))
```

## 실습10(재귀). 두 리스트를 결합

```python
def combine_recursive(l=[1,2,3,4,5,6,7], r=[100,200,300]):
    if not l: return r
    return [l[0]] + combine_recursive(l[1:], r)
print(combine_recursive())
```

## 실습11(꼬리재귀). 문자열 뒤집기

```python
def reverse_tail(l="Hello World", acc=[]):
    if not l: return acc
    return reverse_tail(l[1:], [l[0]] + acc)
print("".join(reverse_tail()))
```

## 실습12(꼬리재귀). 리스트에서 원소 가져오기

```python
def take_tail(l=[1,2,3,4,5,6], n=2, acc=[]):
    if n <= 0 or not l: return acc
    return take_tail(l[1:], n-1, acc + [l[0]])
print(take_tail())
```

## 실습13(꼬리재귀). 두 리스트 결합

```python
def combine_tail(l=[1,2,3,4,5,6], r=[100,200,300], acc=[]):
    if not l and not r:
        return acc
    elif l:
        return combine_tail(l[1:], r, acc + [l[0]])
    else:
        return combine_tail(l, r[1:], acc + [r[0]])
print(combine_tail())
```

# 과제2(a) - 병합 정렬(@knuth1998taocp3)
- 병합 정렬(@knuth1998taocp3)의 `merge(l1,l2)`을 구현해서 제출하세요.
- [Assignment02](https://raw.githubusercontent.com/sigmadream/ds-bootcamp/main/slide/assignment02.py)

```python
data = [3, 0, 1, 8, 7, 2, 5, 4, 6, 9]

def merge(l1, l2):
  i = 0
  j = 0
  merge_list = []
  
  while i < len(l1) and j < len(l2):
    if l1[i] < l2[j]:
      merge_list.append(l1[i])
      i = i + 1
    else:
      merge_list.append(l2[j])
      j = j + 1  
  
  if(i < len(l1)):
    for x in range (i, len(l1)):
      merge_list.append(l1[x])
  
  if(j < len(l2)):
    for x in range (j, len(l2)):
      merge_list.append(l2[x])
  
  return merge_list
def merge_sort(data):
  size = 1
  n = len(data)
  while size < n:
    i1 = 0
    i2 = size
    while i2 < n:
      merge_list = merge(data[i1:i2], data[i2:(i2 + size)])
      i = 0
      x = i1
      while x < i2 + size and i < len(merge_list) :
        data[x] = merge_list[i]
        i = i + 1
        x = x + 1
      i1 = i2 + size
      i2 = i1 + size
    size = size * 2
  return data

merge_list = merge(data[0:len(data)//2], data[len(data)//2:len(data)])
print(merge_list)
merge_sort = merge_sort(data)
print(merge_sort)
```

## 과제2(b) - 병합 정렬(@knuth1998taocp3)

```python
def merge_recursive(left, right):
    if not left:
        return right
    if not right:
        return left
    if left[0] < right[0]:
        return [left[0]] + merge_recursive(left[1:], right)
    return [right[0]] + merge_recursive(left, right[1:])

def merge_sort_recursive(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort_recursive(arr[:mid])
    right = merge_sort_recursive(arr[mid:])
    return merge_recursive(left, right)
```

```python
merge_sort = merge_sort_recursive(data)
print(merge_sort)
```

## 과제2(c) - 병합 정렬(@knuth1998taocp3)

```python
def merge_tail_recursive(left, right, acc):
    if not left:
        return acc + right
    if not right:
        return acc + left
    if left[0] < right[0]:
        return merge_tail_recursive(left[1:], right, acc + [left[0]])
    return merge_tail_recursive(left, right[1:], acc + [right[0]])

def merge_sort_tail_recursive(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort_tail_recursive(arr[:mid])
    right = merge_sort_tail_recursive(arr[mid:])
    return merge_tail_recursive(left, right, [])
```

```python
merge_sort = merge_sort_tail_recursive(data)
print(merge_sort)
```
